```
#File location and type
file_location = "s3a://first-street-climate-risk-statistics-for-noncommercial-use/01_DATA/Climate_Risk_Statistics/OLDER_VERSIONS/v1.1/Zip_level_risk_FEMA_FSF_v1.1.csv"
file_type = "csv"
# CSV options
infer_schema =True
first_row_is_header =True
delimiter = ","
# The applied options are for CSV files . For other file types , these will be ignored .
df = spark.read.format(file_type).option("inferSchema", infer_schema).option("header", first_row_is_header).option("sep", delimiter).load (file_location)
display(df)
permanent_table_name = "zip_level_risk_v1_1"
df.write.saveAsTable(permanent_table_name)
```

```
#File location and type
file_location = "s3a://first-street-climate-risk-statistics-for-noncommercial-use/01_DATA/Climate_Risk_Statistics/OLDER_VERSIONS/v1.2/Zip_level_risk_FEMA_FSF_v1.2.csv"
file_type = "csv"
# CSV options
infer_schema = True
first_row_is_header = True
delimiter = ","
# The applied options are for CSV files . For other file types , these will be ignored .
df = spark.read.format(file_type).option("inferSchema", infer_schema).option("header", first_row_is_header).option("sep", delimiter).load (file_location)
display(df)
permanent_table_name = "zip_level_risk_v1_2"
df.write.saveAsTable(permanent_table_name)
```

```
v1_1DF = spark.table('zip_level_risk_v1_1')
v1_2DF = spark.table('zip_level_risk_v1_2')
```

```
v1_1DF.show()
v1_2DF.show()
```

```
v1DF = v1_1DF.select("zipcode", "count_property").show()
```

```
v2DF = v1_2DF.select("zipcode", "count_property").show()
```

```
v1_1_reducedDF = v1_1DF.select(["zipcode","count_property"]).filter(v1_1DF['zipcode']<=1050).withColumnRenamed("count_property","cp_v1_1").withColumnRenamed("zipcode","zcode_v1_1")
```

```
v1_1_reducedDF.show()
```

```
v1_2_reducedDF = v1_2DF.select(["zipcode","count_property"]).filter(v1_2DF['zipcode']<=1050).withColumnRenamed("count_property","cp_v1_2").withColumnRenamed("zipcode","zcode_v1_2")
```

```
v1_2_reducedDF.show()
```

```
joineddf = v1_1_reducedDF.join(v1_2_reducedDF, v1_1_reducedDF.zcode_v1_1 == v1_2_reducedDF.zcode_v1_2, 'inner')
```

```
joineddf.show()
```

```
joineddf.where(joineddf.zcode_v1_1 == 1020).show()
```

```
x_axis =[]
cp_v1_1 =[]
cp_v1_2 =[]
for row in joineddf.rdd.collect():
    x_axis.append(row.zcode_v1_1)
    cp_v1_1.append(row.cp_v1_1)
    cp_v1_2.append(row.cp_v1_2)
```

```
versions = ["cp_v1_1","cp_v1_2"]
data = {
    "names" : x_axis ,
    "cp_v1_1" : cp_v1_1 ,
    "cp_v1_2" : cp_v1_2
}
```

```
from bokeh.plotting import figure
from bokeh.embed import components, file_html
from bokeh.resources import CDN

#from bokeh.io import output_file, show
#from bokeh.plotting import figure

colors = ["#c9d9d3", "#718dbf"]

p = figure(plot_height =250, title ="Property counts by version",
        toolbar_location=None, tools="hover", tooltips="name@names:@name")


p.vbar_stack(versions, x='names', width= 0.9, color=colors, source=data,
        legend_label=versions)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation ="horizontal"

# create an html document that embeds the Bokeh plot
html =file_html(p, CDN, "my plot1")

#displays html
displayHTML(html)
```
